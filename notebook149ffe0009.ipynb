```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe=pd.read_csv("/kaggle/input/iris-data-set/iris_data (1) - iris_data (1).csv")
dataframe.head(n=10)
```

```python
for x in range(len(dataframe)):
    if dataframe.iloc[x, 2] == "Iris-sentosa":
        dataframe.iloc[x, 2] = 0
    else:
        dataframe.iloc[x, 2] = 1
        
dataframe.head()
```

```python
plt.scatter(dataframe["PetalLengthCm"], dataframe["PetalWidthCm"], c=dataframe["Species"], cmap="Spectral")

plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
cbar = plt.colorbar()
```
